```python
import pandas as pd
#cargar archivo
datos = pd.read_csv("/content/drive/MyDrive/competencia/train.csv", sep=",")
train = pd.DataFrame(datos)
print(train.head())
```

```python
display(train.describe())
```

```python
train.columns = train.columns.str.strip()
train.columns = train.columns.str.replace(' ', '_')
train.columns = train.columns.str.replace('.', '_')
train.columns = train.columns.str.upper()
```

```python
y = train['RENDIMIENTO_GLOBAL']     # tenemos que guardala para que en el paso de codificacion y normalizacion no afecte el modelo
X = train.drop('RENDIMIENTO_GLOBAL', axis=1)
```

# Revision de columnas sin datos o faltantes

```python
train.info()
train.isnull().sum()
```

Debido a que algunas columnas no tiene valores, las columnas que son de categoria, las vamos a llenar con los valores que mas se repiten, es decir la moda.

```python
from sklearn.impute import  SimpleImputer

imputer_num = SimpleImputer(strategy='mean') # Crear imputador
imputer_cte = SimpleImputer(strategy='most_frequent') # Crear imputador

num = X.select_dtypes(include=['int64', 'float64']).columns
cte = X.select_dtypes(include=['object']).columns

X[num] = imputer_num.fit_transform(X[num])
X[cte] = imputer_cte.fit_transform(X[cte])
```

```python
train.isnull().sum()
```

```python
train.head()
```

```python
train.select_dtypes(include=['int64', 'float64']).isnull().sum()[lambda x: x > 0] # verificacion de columnas numericas vacias
```

# Codificación One-Hot

Para que el modelo pueda procesar las columnas de tipo object, que son variables categóricas, las convertimos en variables númericas.
Usamos "drop_first = true" para reducir la multicolinealidad, es decir evitar la duplicidad de información

```python
X = pd.get_dummies(X, drop_first=True)
```

# Normalización

La variable ID no se normaliza debido a que funciona como un identificador, no es una variable con algún sentido matemático para nuestro modelo

```python
from sklearn.preprocessing import MinMaxScaler

column_normalizar = [
    'INDICADOR_1',
    'INDICADOR_2',
    'INDICADOR_3',
    'INDICADOR_4'
]

scaler = MinMaxScaler()
X[column_normalizar] = scaler.fit_transform(X[column_normalizar])
```
